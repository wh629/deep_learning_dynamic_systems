### Multi-run Lorenz

```python
%pylab inline

import numpy as np
import tensorflow as tf
from scipy.integrate import odeint
from utils import *
np.random.seed(0)

#
#
noise_percent = 5
#
#
```

### Generate data

```python
def lorenz_ode(X, t, params):
    
    s,r,b = params
    x,y,z = X

    return [s*(y-x), x*(r-z)-y, x*y-b*z]

M = 20000
subsample_rate = 100

subsample = [j for j in range(M) if j%subsample_rate == 0]
m = len(subsample)
n = 3
num_runs = 50

t = np.linspace(0,2.5,M); dt = (t[1]-t[0])*subsample_rate

params = (10,28,8.0/3)
X_all = [odeint(lorenz_ode, [0,0,25] + 10*np.random.randn(3), t, args=(params,), rtol = 1e-12, atol = 1e-12)[subsample, :].T \
         for _ in range(num_runs)]


figure(figsize = (15,5))

for j in range(num_runs):
    
    x = X_all[j][0,:]; y = X_all[j][1,:]; z = X_all[j][2,:]
    
    subplot(1,3,1); plot(x, y)
    subplot(1,3,2); plot(y, z)
    subplot(1,3,3); plot(x, z)
```

```python
noise_level = [0.01*noise_percent*np.std(X) for X in X_all]
measurement_noise = [noise_level[i]*np.random.randn(n,m) for i in range(num_runs)]

Y_all = [X + N for (X,N) in zip(X_all, measurement_noise)]
```

```python
N_hat_all = [approximate_noise(Y)[0] for Y in Y_all]
```

```python
figure(figsize = (15,5))
subplot(1,3,1); plot(Y_all[0][0,:], 'b')
subplot(1,3,2); plot(Y_all[0][1,:], 'b')
subplot(1,3,3); plot(Y_all[0][0,:],Y_all[0][1,:], 'b')

figure(figsize = (15,5))
subplot(1,3,1); plot(Y_all[0][0,:], 'b'); plot((Y_all[0] - N_hat_all[0])[0,:], 'r')
subplot(1,3,2); plot(Y_all[0][1,:], 'b'); plot((Y_all[0] - N_hat_all[0])[1,:], 'r')
subplot(1,3,3); plot(Y_all[0][0,:],Y_all[0][1,:], 'b')
plot((Y_all[0] - N_hat_all[0])[0,:],(Y_all[0] - N_hat_all[0])[1,:], 'r')
```

### Train 

```python
tf.reset_default_graph()
sess = tf.InteractiveSession()
```

### Create trainable variables

```python
#####################################################################################################
# 
#  Get trainable tensorflow variables for neural network and each matrix for measurement noise
#
#####################################################################################################

n_hidden = 3
size_hidden = 32

#####################################################################################################

layer_sizes = [n] + [size_hidden for _ in range(n_hidden)] + [n]
num_layers = len(layer_sizes)

weights = []
biases = []

for j in range(1,num_layers):
    weights.append(tf.get_variable("W"+str(j), [layer_sizes[j],layer_sizes[j-1]], \
                                   initializer = tf.contrib.layers.xavier_initializer(seed = 1)))
    biases.append(tf.get_variable("b"+str(j), [layer_sizes[j],1], initializer = tf.zeros_initializer()))

N_all = [tf.get_variable("N"+str(i), initializer = N_hat_all[i].astype('float32')) for i in range(num_runs)]
```

### Create placeholders

```python
num_dt = 10

Y_0 = [tf.placeholder(tf.float32, [n,None], name = "Y"+str(i)+"_0") for i in range(num_runs)]
T_0 = tf.placeholder(tf.float32, [1,None], name = "T_0")

true_forward_Y = [[] for _ in range(num_runs)]
true_backward_Y = [[] for _ in range(num_runs)]

for i in range(num_runs):
    for j in range(num_dt):
        true_forward_Y[i].append(tf.placeholder(tf.float32, [n,None], name = "Y"+str(i)+"_"+str(j+1)+"_true"))
        true_backward_Y[i].append(tf.placeholder(tf.float32, [n,None], name = "Y"+str(i)+"_n"+str(j+1)+"_true"))

h = tf.placeholder(tf.float32, [1,1], name = "h")
```

### Forward and backward predictions

```python
method = 'RK4'

X_0 = [tf.subtract(Y_0[i], tf.slice(N_all[i], [0,num_dt],[n,m-2*num_dt])) for i in range(num_runs)]

pred_forward_X = []
pred_backward_X = []

for i in range(num_runs):
    
    pred_forward_X.append([RK_timestepper(X_0[i], T_0, simple_net, h, \
                                          weights, biases, method = method)])
    pred_backward_X.append([RK_timestepper(X_0[i], T_0, simple_net, h, \
                                           weights, biases, method = method, direction = 'B')])

    for j in range(1,num_dt):
        pred_forward_X[i].append(RK_timestepper(pred_forward_X[i][-1], T_0, simple_net, \
                                                h, weights, biases, method = method))
        pred_backward_X[i].append(RK_timestepper(pred_backward_X[i][-1], T_0, simple_net, h, \
                                                 weights, biases, method = method, direction = 'B'))

pred_forward_Y = [[pred_forward_X[i][j] + tf.slice(N_all[i], [0,num_dt+1+j],[n,m-2*num_dt]) \
                  for j in range(num_dt)] for i in range(num_runs)]
        
pred_backward_Y = [[pred_backward_X[i][j] + tf.slice(N_all[i], [0,num_dt-1-j],[n,m-2*num_dt]) \
                  for j in range(num_dt)] for i in range(num_runs)]
```

### Create cost function

```python
beta = 1e-6
gamma = 1e-5
decay_const = 0.9

#################################################################################################

output_weights = [decay_const**j for j in range(num_dt)]

fidelity = []

for i in range(num_runs):

    forward_fidelity = tf.reduce_sum([w*tf.losses.mean_squared_error(true,pred) \
                          for (w,true,pred) in zip(output_weights,true_forward_Y[i],pred_forward_Y[i])])

    backward_fidelity = tf.reduce_sum([w*tf.losses.mean_squared_error(true,pred) \
                          for (w,true,pred) in zip(output_weights,true_backward_Y[i],pred_backward_Y[i])])

    fidelity.append(tf.add(forward_fidelity, backward_fidelity))

# Regularizer for NN weights
weights_regularizer = tf.reduce_mean([tf.nn.l2_loss(W) for W in weights])

# Regularizer for explicit noise term
noise_regularizer = tf.reduce_sum([tf.nn.l2_loss(N) for N in N_all])

# Weighted sum of individual cost functions
cost = tf.reduce_sum(tf.reduce_sum(fidelity) + beta*weights_regularizer + gamma*noise_regularizer)

# BFGS optimizer via scipy
optimizer = tf.contrib.opt.ScipyOptimizerInterface(cost, options={'maxiter': 50000, 
                                                                  'maxfun': 50000,
                                                                  'ftol': 1e-15, 
                                                                  'gtol' : 1e-11,
                                                                  'eps' : 1e-12,
                                                                  'maxls' : 100})
```

```python
init = tf.global_variables_initializer()
sess.run(init)
```

### Run Optimizer

```python
if True: 
    
    saver = tf.train.Saver()
    saver.restore(sess, "./saved_sessions/Lorenz_multi_5.ckpt")
    
else:
    
    feed_dict = {T_0 : (np.arange(m-2*num_dt)*dt).reshape(1,m-2*num_dt),
                h : np.array([[dt]])}

    for i in range(num_runs):

        feed_dict[Y_0[i]] = Y_all[i][:,num_dt:m-num_dt]

        for j in range(num_dt):

            feed_dict[true_forward_Y[i][j]] = Y_all[i][:,num_dt+j+1:m-num_dt+j+1]
            feed_dict[true_backward_Y[i][j]] = Y_all[i][:,num_dt-j-1:m-num_dt-j-1]

    optimizer.minimize(sess, feed_dict=feed_dict)
```

### How does it do on training data?

```python
run_num = 0

X_test = np.zeros((n,m))
X_test[:,0] = X_all[run_num][:,0]

x_next_test = RK_timestepper(Y_0[0], T_0, simple_net, h, weights, biases)

for j in range(m-1):
    X_test[:,j+1] = sess.run(x_next_test, feed_dict={Y_0[0]: np.array([X_test[:,j]]).T, 
                                                    T_0: np.array([[j*dt]]), 
                                                    h: np.array([[dt]])}).flatten()
    
figure(figsize = (15,5))
subplot(1,3,1); plot(X_all[run_num][0,:], 'b'); plot(X_test[0,:], 'r')
subplot(1,3,2); plot(X_all[run_num][1,:], 'b'); plot(X_test[1,:], 'r')
subplot(1,3,3); plot(X_all[run_num][0,:],X_all[run_num][1,:], 'b'); plot(X_test[0,:],X_test[1,:], 'r')

figure(figsize = (15,5))
subplot(1,3,1); plot(Y_all[run_num][0,:], 'b'); plot(X_test[0,:], 'r')
subplot(1,3,2); plot(Y_all[run_num][1,:], 'b'); plot(X_test[1,:], 'r')
subplot(1,3,3); plot(Y_all[run_num][0,:],Y_all[run_num][1,:], 'b'); plot(X_test[0,:],X_test[1,:], 'r')

start = 0
end = m

figure(figsize = (15,8))

plot(measurement_noise[run_num][0,start:end], label = 'Measurement noise')
plot(N_all[run_num].eval()[0,start:end], label = 'Learned measurement noise')
plot(measurement_noise[run_num][0,start:end] - N_all[run_num].eval()[0,start:end], 'r', label = 'Noise error')
legend(loc='upper right')
```
